```python
agg_functions = {'calories':'first','favoriet': 'max', 'aanbevolen': 'max', 'vis': 'max', 'vega': 'max'} #rijen die behoren tot één recept samenvoegen
new_data_rec = new_data_rec.groupby(new_data_rec['title']).aggregate(agg_functions).reset_index() #de index resetten
new_data_rec
```

```python
CalProbleem = LpProblem('Calorieën_minimaliseren', LpMinimize)
```

```python
# parameters
n = len(new_data_rec) # Aantal recepten
c = new_data_rec['calories'].tolist() # Aantal calorieën per recept
f = new_data_rec['favoriet'].tolist() # Of recept een favoriet recept is
a = new_data_rec['aanbevolen'].tolist() # Of recept een aanbevolen recept is
vi = new_data_rec['vis'].tolist() # Of recept een vis gerecht is
ve = new_data_rec['vega'].tolist() # Of recept een vega gerecht is
M = 1000  # grote waarde voor hulpvariabelen

# Beslissingsvariabelen
x = LpVariable.dicts("Recept", range(n), cat='Binary')  # Beslissingsvariabelen x_i
z = LpVariable('z', cat='Binary')

# Doelfunctie: Minimaliseren van het aantal calorieën
CalProbleem += lpSum(c[i] * x[i] for i in range(n))

# Restricties
CalProbleem += lpSum(x[i] for i in range(n)) == 7  # 7 recepten
CalProbleem += lpSum(c[i] * x[i] for i in range(n)) >= 600 * 7  # Min aantal calorieën
CalProbleem += lpSum(c[i] * x[i] for i in range(n)) <= 875 * 7  # Max aantal calorieën
CalProbleem += lpSum(f[i] * x[i] for i in range(n)) == 4  # 4 favoriete recepten
CalProbleem += lpSum(a[i] * x[i] for i in range(n)) == 3  # 3 aanbevolen recepten
CalProbleem += lpSum(vi[i] * x[i] for i in range(n)) >= 2 + M*z  # 2 visgerechten
CalProbleem += lpSum(ve[i] * x[i] for i in range(n)) >= 1 + M * (1 - z)  # 1 vega gerecht
```

```python
print(CalProbleem)
```

```python
status = CalProbleem.solve()
print(LpStatus[CalProbleem.status])
```
