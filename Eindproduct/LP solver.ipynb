```python
CalProbleem = LpProblem('Calorieën_minimaliseren', LpMinimize)
# parameters
rec_title = new_data_rec['recipe'].tolist()

n = len(new_data_rec) # Aantal recepten
vet = new_data_rec['vet'].tolist() # Aantal vetten per recept
c = new_data_rec['energie'].tolist() # Aantal calorieën per recept
f = new_data_rec['favoriet'].tolist() # Of recept een favoriet recept is
a = new_data_rec['aanbevolen'].tolist() # Of recept een aanbevolen recept is
vi = new_data_rec['vis'].tolist() # Of recept een vis gerecht is
ve = new_data_rec['vega'].tolist() # Of recept een vega gerecht is
M = 1000  # grote waarde voor hulpvariabelen

# Beslissingsvariabelen
#x = LpVariable.dicts("Recept", range(n), cat='Binary')  # Beslissingsvariabelen x_i
x = LpVariable.dicts("Recept", rec_title, cat='Binary')  # Beslissingsvariabelen x_i
z = LpVariable('z', cat='Binary')
# z_vis = LpVariable('z_vis', cat='Binary')
# z_vega = LpVariable('z_vega', cat='Binary')

# Doelfunctie: Minimaliseren van het aantal vetten
CalProbleem += lpSum(vet[i] * x[rec_title[i]] for i in range(n))


# Restricties
CalProbleem += lpSum(x[rec_title[i]] for i in range(n)) == 7, "Aantal_recepten"  # 7 recepten
#CalProbleem += lpSum(c[i] * x[rec_title[i]] for i in range(n)) >= 600 * 7  # Min aantal calorieën
#CalProbleem += lpSum(c[i] * x[rec_title[i]] for i in range(n)) <= 875 * 7  # Max aantal calorieën
CalProbleem += lpSum(f[i] * x[rec_title[i]] for i in range(n)) == 4, "Aantal_fav"  # 4 favoriete recepten
CalProbleem += lpSum(a[i] * x[rec_title[i]] for i in range(n)) == 3, "Aantal_aanb"  # 3 aanbevolen recepten
#CalProbleem += z_vis + z_vega == 1, "Vis_Or_Vega"
CalProbleem += lpSum(vi[i] * x[rec_title[i]] for i in range(n)) >= 2 - M*z, "Vis"  # 2 visgerechten
CalProbleem += lpSum(ve[i] * x[rec_title[i]] for i in range(n)) >= 1 - M*(1-z), "Vega"  # 1 vega gerecht



# Caloriebeperkingen voor elk recept individueel
for i in range(n):
     CalProbleem += x[rec_title[i]] * (c[i]-600 ) >= 0, f"Min_Calories_{rec_title[i]}"
     CalProbleem += x[rec_title[i]] * (c[i]-875) <= 0, f"Max_Calories_{rec_title[i]}"
```
