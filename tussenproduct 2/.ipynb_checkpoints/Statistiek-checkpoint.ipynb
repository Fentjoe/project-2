```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

ingr_schoon=pd.read_csv('tabellen/data_ingr_clean.csv')
gr_dan =  ingr_schoon.copy()[(ingr_schoon["quantity"] > 60000)]
meest_voorkomend = gr_dan["ingredient"].value_counts().index.tolist()
tabel_ds= gr_dan.loc[gr_dan['ingredient'].isin(meest_voorkomend)]
ingr_schoon[ingr_schoon['ingredient']=='azijn']
gr_dan
gr_dan["ingredient"].value_counts()
```

```python
plt.scatter(tabel_ds['quantity'], tabel_ds['ingredient'])
plt.xticks(rotation = 90, ha='right')
plt.grid()
plt.show()
```

```python
#mean_lijst = []
#gr_dan2 = pd.DataFrame()
for i in gr_dan['ingredient']:   
    median_i = ingr_schoon['quantity'].loc[ingr_schoon['ingredient']== i].median()
    #ingr_schoon['quantity'].loc[ingr_schoon["ingredient"] == i['ingredient'], 'ingredient'] = median_i
    gr_dan['quantity'].mask(gr_dan['ingredient'] == i, median_i, inplace=True)
    #gr_dan2.append(gr_dan['quantity'].loc[gr_dan['ingredient']==i] = median_i)
       
gr_dan

#ingr_schoon.loc[ingr_schoon['ingredient']=='azijn']
```
